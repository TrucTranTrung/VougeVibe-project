```python
import os

# Đường dẫn tới folder chứa các file .txt
folder_path = "C:/Users/ASUS/Downloads/total.v2i.yolov9/train/labels"

# Định nghĩa class cần thay đổi
class_to_change = 10
correct_class = 9

# Duyệt qua tất cả các file .txt trong folder
for filename in os.listdir(folder_path):
    if filename.endswith(".txt"):
        file_path = os.path.join(folder_path, filename)
        
        # Đọc file và thay thế class
        with open(file_path, 'r') as file:
            lines = file.readlines()
        
        new_lines = []
        for line in lines:
            parts = line.split()
            # Kiểm tra nếu class đầu tiên là class_to_change thì thay thế nó
            if int(parts[0]) == correct_class:
                print(f"File {filename} có class đầu tiên là {class_to_change}")
                break

print("Hoàn thành việc thay thế class.")
```
